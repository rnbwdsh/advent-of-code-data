````python
from pathlib import Path
import json
import re

cnt = 0
total = 0
for folder in Path("example_data").glob("*"):
    for file in folder.glob("*"):
        try:
            fc = open(file).read()
            fr = json.loads(fc)
            assert "input" in fr and "output" in fr
            cnt += 1
        except Exception as e1:
            try:
                start_end = [(i.start(), i.end()) for i in re.finditer("```", fc)]
                if len(start_end) >= 2:
                    fc = fc[start_end[0][1]: start_end[1][0]]
                    if fc.startswith("json"):
                        fc = fc[4:].strip()
                    fr = json.loads(fc)
                    cnt += 1
                else:
                    print(file, e1)
                    print(fc)
                    print("-"*50)
                    raise
            except Exception as e2:
                e1 = e2
                
        total += 1
cnt/total
````
